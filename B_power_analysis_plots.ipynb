```python
import numpy as np
import pandas as pd
import scipy.stats
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import seaborn
import pingouin as pg
from scipy.stats import multivariate_normal as mvn
```

For more liberal power analysis: supplementary figure 1

```python
mixed_effects_power_table=pd.read_csv('csvs/liberal_power_table.csv')
```

```python
mixed_effects_power_table.reset_index(inplace=True)
for i in mixed_effects_power_table.index:
    power=mixed_effects_power_table.iloc[i,:].power
    if power >= 0.95:
        mixed_effects_min_subjects=mixed_effects_power_table.iloc[i,:].num_subjects
        break
mixed_effects_min_subjects
```

```python
plt.scatter(mixed_effects_power_table['num_subjects'], mixed_effects_power_table['power'], s=7)
plt.axhline(y=0.95, color='r')
plt.axvline(x=mixed_effects_min_subjects, color='g')
plt.suptitle("Medium Effect Size Power Analysis", fontsize=12, fontname="Arial")
plt.title("Alpha=0.05, Effect size=small (Cohen's d=0.2)", fontsize=10, fontname="Arial")
plt.xlabel("Number of subjects", fontsize=10, fontname="Arial")
plt.xticks(fontsize=9, fontname="Arial")
plt.yticks(fontsize=9, fontname="Arial")
plt.ylabel("Power", fontsize=10, fontname="Arial")
plt.annotate('Power=95%', xy=(5,0.96), color='red', fontname="Arial", fontsize=9)
annotation='Min subjects = '+str(mixed_effects_min_subjects)
plt.annotate(annotation, xy=(mixed_effects_min_subjects+2,0.23), color='green', fontname="Arial", fontsize=9)
#plt.savefig("BagleySupplementaryFig1_revised.pdf")
#plt.savefig("BagleySupplementaryFig1_revised.png", dpi=1200)
```
